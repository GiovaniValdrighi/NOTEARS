```python
import sys
sys.path.append('../src/')
from linear import notears_linear
from nonlinear import notears_nonlinear
from utils import simulate_dag, simulate_nonlinear_sem, simulate_linear_sem

import matplotlib.pyplot as plt
import numpy as np
import networkx as nx
```

## Using the linear model
We simulate a random DAG with 5 variables and we'll use a sample of 200 observations.

```python
linear_DAG = simulate_dag(5, 5)
linear_df = simulate_linear_sem(linear_DAG, 200)

est_linear_DAG = notears_linear(linear_df, lambda1 = 0.01)
```

```python
c_max = np.ceil(max(np.max(linear_DAG), np.max(est_linear_DAG)))
c_min = np.floor(min(np.min(linear_DAG), np.min(est_linear_DAG)))
fig, axis = plt.subplots(1, 2)
fig.set_figwidth(10)
fig.set_figheight(4)

axis[0].imshow(linear_DAG, vmax = c_max, vmin = c_min, cmap = 'viridis')
im = axis[1].imshow(est_linear_DAG, vmax = c_max, vmin = c_min, cmap = 'viridis')

fig.colorbar(im, ax=axis.ravel().tolist())
plt.show()
```

Also with a bigger DAG and dataset.

```python
linear_DAG_big = simulate_dag(20, 16)
linear_df_big = simulate_linear_sem(linear_DAG_big, 1000)

est_linear_DAG_big = notears_linear(linear_df_big, lambda1 = 0.01)
```

```python
c_max = np.ceil(max(np.max(linear_DAG_big), np.max(est_linear_DAG_big)))
c_min = np.floor(min(np.min(linear_DAG_big), np.min(est_linear_DAG_big)))
fig, axis = plt.subplots(1, 2)
fig.set_figwidth(10)
fig.set_figheight(4)

axis[0].imshow(linear_DAG_big, vmax = c_max, vmin = c_min, cmap = 'viridis')
im = axis[1].imshow(est_linear_DAG_big, vmax = c_max, vmin = c_min, cmap = 'viridis')

fig.colorbar(im, ax=axis.ravel().tolist())
plt.show()
```

## Using the nonlinear model

We simulate a random DAG with 5 variables using the Erdos Renvy method, next we are using a sample generated from this DAG of 200 observations.

```python
nonlinear_DAG = simulate_dag(5, 9)
nonlinear_df = simulate_nonlinear_sem(nonlinear_DAG, 200).astype(np.float32)

est_DAG = notears_nonlinear([5, 10, 1], nonlinear_df, lambda1 = 0.01, lambda2 = 0.01)
```

```python
c_max = np.ceil(max(np.max(nonlinear_DAG), np.max(est_DAG)))
c_min = np.floor(min(np.min(nonlinear_DAG), np.min(est_DAG)))
fig, axis = plt.subplots(1, 2)
fig.set_figwidth(10)
fig.set_figheight(4)

axis[0].imshow(nonlinear_DAG, vmax = c_max, vmin = c_min, cmap = 'viridis')
im = axis[1].imshow(est_DAG, vmax = c_max, vmin = c_min, cmap = 'viridis')

fig.colorbar(im, ax=axis.ravel().tolist())
plt.show()
```

We can also use with a bigger DAG with 20 variables, in this situation we'll simulate 2000 samples.

```python
nonlinear_DAG_big = simulate_dag(20, 16)
nonlinear_df_big = simulate_nonlinear_sem(nonlinear_DAG_big, 2000).astype(np.float32)

est_DAG_big = notears_nonlinear([20, 30, 20, 1], nonlinear_df_big, lambda1 = 0.01, lambda2 = 0.01)
```

```python
c_max = np.ceil(max(np.max(nonlinear_DAG_big), np.max(est_DAG_big)))
c_min = np.floor(min(np.min(nonlinear_DAG_big), np.min(est_DAG_big)))
fig, axis = plt.subplots(1, 2)
fig.set_figwidth(10)
fig.set_figheight(4)

axis[0].imshow(nonlinear_DAG_big, vmax = c_max, vmin = c_min, cmap = 'viridis')
im = axis[1].imshow(est_DAG_big, vmax = c_max, vmin = c_min, cmap = 'viridis')

fig.colorbar(im, ax=axis.ravel().tolist())
plt.show()
```
